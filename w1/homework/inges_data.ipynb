```python
from datetime import datetime

import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('green_tripdata_2019-01.csv.gz')
df.head()
```

```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
date_start = pd.to_datetime('2019-01-15 00:00:00')
date_finish = pd.to_datetime('2019-01-15 23:59:59')
df.loc[(df.lpep_pickup_datetime >=date_start) & (df.lpep_dropoff_datetime <= date_finish)]
```

```python
df.loc[ lambda df: df['trip_distance']== max(df.trip_distance)]
```

```python
date_start_5 = pd.to_datetime('2019-01-01 00:00:00')
date_finish_5 = pd.to_datetime('2019-01-01 23:59:59')

df.loc[
    (df['lpep_pickup_datetime'] >= date_start_5) &
    (df['lpep_dropoff_datetime'] <= date_finish_5) &
    (df['passenger_count'] == 3)
    ]
```

```python
df[df['PULocationID']==7].loc[lambda df: df['trip_distance'] == max(df['trip_distance'])]
```
